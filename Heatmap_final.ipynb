## 1. load in data

```python
import pandas as pd
data = pd.read_csv(r'E:\WEMUST\Opensourcetools\final\counts_final.csv')
data.head()
```

## 2. Generate two-dimensional lists in order of Country names

```python
countryname=data.columns[1:]
print(countryname)
flow=[]
sj=list(data['Year'])
#print(sj)
for i in range(len(sj)):
    flow.append([])
    for j in range(len(countryname)):
        flow[i].append(int(list(data[countryname[j]])[i]))
#print(flow)
```

## 3. Generate Heatmaps by year

```python
from pyecharts import options as opts
from pyecharts.charts import Map, Timeline
from pyecharts.faker import Faker
```

```python
#set timeline and playback mode ( is_auto_play = False means play manually, if play automatically, the time interval is 500 ms,)
timeline = Timeline(init_opts=opts.InitOpts(width='1400px', height='600px')) 
timeline.add_schema(is_auto_play=False, play_interval=500) 

#support chain calls, here use traditional way
for m in range(len(sj)):
    c=(Map(init_opts=opts.InitOpts(width="1400px",height='600px',bg_color="white"))) #initialize figure size  
    c.add("", [list(z) for z in zip(countryname, flow[m])], "world",is_map_symbol_show=False) #generate map
    c.set_series_opts(label_opts=opts.LabelOpts(is_show=False), # label "do not show country's name"
                      tooltip_opts=opts.TooltipOpts(is_show=True,formatter='{b} : {c}'))  #set tip tool format, {b} stands for country name, {c} stands for its vallue
    c.set_global_opts(title_opts=opts.TitleOpts(title="The Number of terrorist attacks in {}".format(sj[m])),   # set title, change with time
        visualmap_opts=opts.VisualMapOpts(max_=500)) #set the maximum number displayed in map
    timeline.add(chart=c,time_point=sj[m]) #add map in timeline

#output the heatmaps with timeline in html format
timeline.render("number_terrorism2.html") 
```
